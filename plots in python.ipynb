```python
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset("iris")

# Draw a box plot using Matplotlib
plt.boxplot([iris['sepal_length'], iris['sepal_width'], iris['petal_length'], iris['petal_width']], labels=iris.columns[:-1])
plt.xlabel('Features')
plt.ylabel('Length/Width (cm)')
plt.title('Box plot of Iris dataset')
plt.show()
```

```python

# Draw a box plot using Matplotlib
ax=iris.boxplot(column='sepal_length',showmeans=True,meanline=True,grid=False)
plt.xlabel('Features')
plt.ylabel('Length/Width (cm)')
plt.title('Box plot of Iris dataset')
plt.show()
```

```python
ax=iris.boxplot(column='sepal_length',by='species',showmeans=True,meanline=True,grid=False)
plt.xlabel('Features')
plt.ylabel('Length/Width (cm)')
plt.title('Box plot of Iris dataset')
plt.show()
```

```python
sns.boxplot(data=iris,x='species',y='sepal_length',hue='species',width=0.3)
plt.show()
```
