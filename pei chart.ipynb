```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore","use_inf_as_na")

iris = sns.load_dataset('iris')
print(iris)
```

```python
import pandas as pd

# Assuming you've already created the DataFrame `iris` with the columns `sepal_length` and `sepal_width`

iris["sepal_length_class"] = pd.cut(iris.sepal_length, bins=[0, 5, 6.5, 10], labels=["short", "medium", "long"])
iris["sepal_width_class"] = pd.cut(iris.sepal_width, bins=[0, 2.7, 6.5, 8], labels=["thin", "normal", "thick"])

iris_tab_1 = pd.crosstab(iris['sepal_length_class'], iris['sepal_width_class'])
iris_tab_2 = iris_tab_1.transpose()

iris_tab_1['all flower'] = iris_tab_1.sum(axis=1)
iris_tab_2['all flower'] = iris_tab_2.sum(axis=1)

# Print the categories of the column 'sepal_length_class'

```

```python
explode = [0.1 if label == 'short' else 0 for label in iris['sepal_length_class']]
ax=iris.plot.pie(y="petal_width" ,explode=explode)
ax.set_title("pie chart")
ax.set_ylabel("all flower values")
plt.legend(loc="upper left")
plt.show()
```

```python
ax=iris_tab_1.plot.pie(y="all flower")
ax.set_title("pie chart")
ax.set_ylabel("all flower values")
plt.legend(loc="upper left")
plt.show()
```

```python
ax=iris_tab_2.plot.pie(y="all flower")
ax.set_title("pie chart 2")
ax.set_ylabel("all flower values chart 2")
plt.legend(loc="upper left")
plt.show()
```
