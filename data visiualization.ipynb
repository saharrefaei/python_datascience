```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore","use_inf_as_na")

iris = sns.load_dataset('iris')
iris2 = pd.get_dummies(iris , columns = ["sepal_width"] , dtype=bool)
print(iris2)
```

```python
plt.figure(figsize=(8, 6))
plt.hist(iris['sepal_length'], bins=10, edgecolor='black',color='skyblue') # bins =به تعداد بخش‌هایی اشاره دارد که داده‌ها به آن‌ها تقسیم می‌شوند.
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', color='grey', alpha=0.7)
plt.show()
```

```python
fix , ax = plt.subplots()
colors = {"setosa":"blue", "versicolor" :"red" , "virginica" : "black" }
for Spieces , data in iris.groupby("species"):
    data.plot(kind='hist',column="sepal_width",bins=15,alpha=0.50,color=colors[Spieces],ax=ax)


plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.grid(True, linestyle='--', color='grey', alpha=0.7)
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), shadow=True, ncol=3)
plt.show()
```

```python
sns.histplot(data=iris,x='sepal_width',y="sepal_length",bins=25,hue="species",element="step")
plt.show()
```

```python

graph = sns.JointGrid()
sns.histplot(data=iris, x='sepal_width', y='sepal_length', bins=25, hue='species', ax=graph.ax_joint)#روی محور اصلی
sns.histplot(data=iris, x='sepal_width', bins=25, hue='species', ax=graph.ax_marg_x)#روی محور ایکس قرار میگیرد
sns.histplot(data=iris, x='sepal_length', bins=25, hue='species', ax=graph.ax_marg_y)#روی محور ایگرگ قرار میگیرد
graph.fig.subplots_adjust(top=0.90)


plt.show()
```
