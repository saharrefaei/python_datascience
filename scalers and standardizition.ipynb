```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

geyser = sns.load_dataset("geyser")
print(geyser)
```

```python
geyser['mc_duration']=geyser['duration']-geyser['duration'].mean() #add new column with name min_duration and then minus the each deuration rows with the mean
geyser['mc_waiting']=geyser['waiting']-geyser['waiting'].mean() #add new column with name min_duration and then minus the each deuration rows with the mean

print(geyser)  
```

```python
print(geyser['duration'].std())  #calculate Standard Deviation
```

```python
geyser['norm_duration']=(geyser['duration']-geyser['duration'].mean())/geyser['duration'].std() #Standardization
geyser['norm_waiting']=(geyser['waiting']-geyser['waiting'].mean())/geyser['waiting'].std() #Standardization

print(geyser) #میانگین آنها صفر و انحراف معیار آنها یک 
#در اینجا یعنی ایندکس 0 ، مقدار دیورایشن اش 0.098318 انحراف بالا تر از میانگین دارد .
#in this case mean the duration of index 0 have 0.098318 standard deviation above the mean
print(geyser['duration'].mean())
```

```python
geyser['norm_waiting']=(geyser['duration']-geyser['duration'].mean())/geyser['duration'].std()
print(geyser)
```

```python
geyser['Z1_duration']=(geyser['duration']-geyser['duration'].min())/(geyser['duration'].max()-geyser['duration'].min())
geyser['Z1_waiting']=(geyser['waiting']-geyser['waiting'].min())/(geyser['waiting'].max()-geyser['waiting'].min())

print(geyser)#Min-Max Scaling
```

```python
fig,((orig,mc),(norm,Z1))=plt.subplots(2,2)
geyser.plot("duration","waiting", kind = "scatter" , ax = orig)
geyser.plot("mc_duration","mc_waiting", kind = "scatter" , ax = mc)
geyser.plot("norm_duration","norm_waiting", kind = "scatter" , ax = norm)
geyser.plot("Z1_duration","Z1_waiting", kind = "scatter" , ax = Z1)

plt.show()
```
