```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

gayser= sns.load_dataset('geyser')
print(gayser)
```

```python
sns.histplot(data=gayser,x='duration',bins=25,hue='kind')
plt.show()
sns.histplot(data=gayser,x='waiting',bins=25,hue='kind')
plt.show()
```

```python
gayser['quc_duration2']= pd.qcut(gayser.duration,q=2,labels=['short','long'])
gayser['quc_duration3']= pd.qcut(gayser.duration,q=3,labels=['short','long','medium'])
print(gayser)
```

```python
fig , ((h_dur,q_dur2,q_dur3)) = plt.subplots(1,3)
sns.histplot(data=gayser,x='duration',bins=26,hue='kind',ax=h_dur)
sns.histplot(data=gayser,x='duration',bins=26,hue='quc_duration2',ax=q_dur2)
sns.histplot(data=gayser,x='duration',bins=26,hue='quc_duration3',ax=q_dur3)

plt.show()
```

```python
gayser['cut_duration2']= pd.cut(gayser.duration,bins=2,labels=['short','long'])
gayser['cut_duration3']= pd.cut(gayser.duration,bins=3,labels=['short','long','medium'])
print(gayser)
```

```python
fig , ((h_dur,q_dur2,q_dur3),(h_dur2,cut_dur2,cut_dur3)) = plt.subplots(2,3)
sns.histplot(data=gayser,x='duration',bins=26,hue='kind',ax=h_dur)
sns.histplot(data=gayser,x='duration',bins=26,hue='quc_duration2',ax=q_dur2)
sns.histplot(data=gayser,x='duration',bins=26,hue='quc_duration3',ax=q_dur3)
sns.histplot(data=gayser,x='duration',bins=26,hue='kind',ax=h_dur2)
sns.histplot(data=gayser,x='duration',bins=26,hue='cut_duration2',ax=cut_dur2)
sns.histplot(data=gayser,x='duration',bins=26,hue='cut_duration3',ax=cut_dur3)
plt.show()
```

```python
gayser['sel_cut_duration']= pd.cut(gayser.duration,bins=[0.3,3,100],labels=['short','long'])
gayser['sel_cut_duration2']= pd.cut(gayser.duration,bins=[0,5,6.6,100],labels=['short','long','medium'])
print(gayser)
```

```python
fig , ((h_dur,q_dur2,q_dur3),(h_dur2,cut_dur2,cut_dur3),(h_dur3,sel_cut_dur2,sel_cut_dur3)) = plt.subplots(3,3)
sns.histplot(data=gayser,x='duration',bins=26,hue='kind',ax=h_dur,legend=None)
sns.histplot(data=gayser,x='duration',bins=26,hue='quc_duration2',ax=q_dur2,legend=None)
sns.histplot(data=gayser,x='duration',bins=26,hue='quc_duration3',ax=q_dur3,legend=None)
sns.histplot(data=gayser,x='duration',bins=26,hue='kind',ax=h_dur2,legend=None)
sns.histplot(data=gayser,x='duration',bins=26,hue='cut_duration2',ax=cut_dur2,legend=None)
sns.histplot(data=gayser,x='duration',bins=26,hue='cut_duration3',ax=cut_dur3,legend=None)
sns.histplot(data=gayser,x='duration',bins=26,hue='kind',ax=h_dur3,legend=None)
sns.histplot(data=gayser,x='duration',bins=26,hue='sel_cut_duration',ax=sel_cut_dur2,legend=None)
sns.histplot(data=gayser,x='duration',bins=26,hue='sel_cut_duration2',ax=sel_cut_dur3,legend=None) #legend = None => prevent show the help button (short,long,medium)
plt.show()
```
